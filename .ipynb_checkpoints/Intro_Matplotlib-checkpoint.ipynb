```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axis = fig.add_axes([0.5,0.1,0.8,0.8]) # controls the left, bottom , width, height of canvas
axis.plot(x,y,'r')
```

Drawing subgraphs

```python
fig, axes = plt.subplots(nrows =1, ncols =2)
for ax in axes:
    ax.plot(x,y,'r')
```

Can also draw graph in a another graph

```python
fig = plt.figure()
axes1 = fig.add_axes([0.5,0.1,0.8,0.8]) # big axes
axes2 = fig.add_axes([0.2,0.5,0.2,0.2]) # small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

# Styling

```python
fig,axes = plt.subplots(dpi = 150)
axes.plot(x,x**2, color = 'red',alpha = 0.5)
axes.plot(x,x+2, color = 'blue',alpha = 0.5)
axes.plot(x,x+3, color = 'green',alpha = 0.5)
```

```python
fig,axes = plt.subplots(dpi = 100)
axes.plot(x,x+1, color = 'blue',linewidth = 0.25)
axes.plot(x,x+2, color = 'blue',linewidth = 1)
axes.plot(x,x+3, color = 'blue',linewidth = 1.5)
axes.plot(x,x+4, color = 'blue',linewidth = 2)
```

```python
fig,axes = plt.subplots(dpi = 100)
axes.plot(x,x+1, color = 'blue',lw = 2, linestyle ='-')
axes.plot(x,x+2, color = 'blue',lw = 2, linestyle = '-.')
axes.plot(x,x+3, color = 'blue',lw = 2, linestyle = ':')

line, = axes.plot(x,x+4,color = 'black',lw=1.50)
line.set_dashes([5,10,15,10])

line, = axes.plot(x,x+5,color = 'black',lw=1.50)
line.set_dashes([5,10,15,3])

line, = axes.plot(x,x+6,color = 'black',lw=1.50)
line.set_dashes([5,10,4,10])
```

```python
fig,axes = plt.subplots(dpi = 100)
axes.plot(x,x+1, color = 'blue',marker = 'o')
axes.plot(x,x+2, color = 'blue', marker = '+')
axes.plot(x,x+3, color = 'blue', marker = 's')
axes.plot(x,x+4, color = 'blue', marker = '1')
axes.plot(x,x+5, color = 'blue',marker = 'o', markersize = 5)
axes.plot(x,x+6, color = 'blue', marker = 's',markerfacecolor= 'red')
```

# Limit -> Limiting Data and 2D graphs


```python
fig, Axes = plt.subplots(1,2,figsize=(10,5))

Axes[0].plot(x,x**2,x**3, lw =2)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x**3, lw =2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)
```

```python
n = np.array([0,1,2,3,4,5])
fig,axes = plt.subplots(1,4,figsize=(16,5))

axes[0].set_title('scatter plot')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title('step plot ')
axes[1].step(n,n**2)

axes[2].set_title('bar plot')
axes[2].bar(n,n**2,align = 'center',width = 0.5,alpha = 0.3)

axes[3].set_title('fill_between')
axes[3].fill_between(x,x**2,x**3,color = 'green',alpha = 0.3)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0,0.0,1,1], polar = True)
t = np.linspace(0,2*np.pi,100)
ax.plot(t,t*5,color='blue',lw=3)
```

```python
n = np.random.randn(100000)
fig,axes = plt.subplots(1,2,figsize=(12,4))
axes[0].set_title('histogram')
axes[0].hist(n,width = .5)

axes[1].set_title('histogram')
axes[1].hist(n,cumulative = True, bins = 50, width = .5)
```

# Drawing Contour image

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1-Z2) * 2
```

```python
fig,ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline = 1, fontsize = 10)
ax.set_title('Contour map')
```

# Drawing a 3d Surface Image

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize = (14,6))

ax= fig.add_subplot(1,1,1,projection = '3d')
ax.plot_surface(X,Y,Z, rstride = 4, cstride = 4, linewidth = 0)
```

## Write a Python Programming to create pie chart of the populatriy of programming languages

```python
import matplotlib.pyplot as plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#','C++'
popularity = [22.2,17.6,8.8,8,7.7,6.7]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
```

```python
explode = (0.1,0,0,0,0,0)
plt.pie(popularity, explode =explode, labels = languages, colors = colors,
autopct='%1.1f%%', shadow = True, startangle = 90)
```
